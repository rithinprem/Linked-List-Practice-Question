```python
#Q1. Define a doubly linked list

class Node:
    def __init__(self, data):
        self._data = data
        self._next = None
        self._prev = None

    def get_data(self):
        return self._data

    def set_data(self, data):
        self._data = data

    def get_next(self):
        return self._next

    def set_next(self, next_node):
        self._next = next_node

    def get_prev(self):
        return self._prev

    def set_prev(self, prev_node):
        self._prev = prev_node


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.get_next():
                current = current.get_next()
            current.set_next(new_node)
            new_node.set_prev(current)

    def prepend(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            new_node.set_next(self.head)
            self.head.set_prev(new_node)
            self.head = new_node

    def display_forward(self):
        current = self.head
        while current:
            print(current.get_data(), end=" <-> ")
            current = current.get_next()
        print("None")

    def display_backward(self):
        current = self.head
        while current.get_next():
            current = current.get_next()
        while current:
            print(current.get_data(), end=" <-> ")
            current = current.get_prev()
        print("None")

# Example usage:
dll = DoublyLinkedList()
dll.append(1)
dll.append(2)
dll.append(3)
dll.prepend(0)

print("Doubly Linked List (Forward):")
dll.display_forward()

print("\nDoubly Linked List (Backward):")
dll.display_backward()
```

```python
# creating linked list

class Node:
  def __init__(self,data,next=None):
    self.data=data
    self.next=next

  def getData(self):
    return self.data

  def setData(self,data):
    self.data=data

  def getNext(self):
    return self.next

  def setNext(self,node):
    self.next=node


def traverse(head):
  temp=head
  while temp:
    print(temp.getData(),end="->")
    temp=temp.getNext()
  print()
```

```python
#Q2. Write a function to reverse a linked list in place


def reverse_ll(head):
  curr=head
  prev=None
  while curr:
    temp=curr.getNext()
    curr.setNext(prev)
    prev=curr
    curr=temp

  return prev


head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)

head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)

traverse(head)

head = reverse_ll(head)

traverse(head)
```

```python
#Q3.Detect cycle in linked list

def detect_cycle(head):
  slow=head
  fast=head

  while slow and fast.getNext():
    slow=slow.getNext()
    fast=fast.getNext().getNext()
    if slow==fast:
      return "Cycle Detected"
  return "No Cylce Detected"


head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)

head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node2)


print(detect_cycle(head))
```

```python
#Q6. Remove duplicates from sorted list

# 1->2->3->3->4->4->4->5  should be changed to 1->2->3->4->5


head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(3)
node4 = Node(4)
node5 = Node(4)
node6 = Node(4)
node7 = Node(5)

head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node5)
node5.setNext(node6)
node6.setNext(node7)

print('LinkedList with duplicates')
traverse(head)

def remove_duplicates(head):
  temp=head
  while temp.getNext():

    if temp.getData()==temp.getNext().getData():
      temp.setNext(temp.getNext().getNext())

    else:
      temp=temp.getNext()

  return head

head=remove_duplicates(head)

print('LinkedList without duplicates')
traverse(head)
```

```python
#Q7. Find the intersection of two linked lists

# 1->2->3->4->8->6->9  5->1->6->7  , intersection 1->6


def intersection(head1,head2):
  temp1=head1
  temp2=head2
  set1 = set()
  headnode= None

  while temp1:

    data= temp1.getData()
    set1.add(data)
    temp1=temp1.getNext()

  while temp2:

    if temp2.getData() in set1:
      if not headnode:
        headnode = Node(temp2.getData())
        temp = headnode
      else:
        node= Node(temp2.getData())
        temp.setNext(node)
        temp = temp.getNext()

    temp2=temp2.getNext()

  return headnode



head1 = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(8)
node5 = Node(6)
node6 = Node(9)

head1.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node5)
node5.setNext(node6)

print('LinkedList1:')
traverse(head1)

head2 = Node(5)
node_1 = Node(1)
node_2 = Node(6)
node_3 = Node(7)

head2.setNext(node_1)
node_1.setNext(node_2)
node_2.setNext(node_3)

print('LinkedList2:')
traverse(head2)

node_head = intersection(head1,head2)
print('Intersection:')
traverse(node_head)
```

```python
#Q8. Rotate a linked list by k positions to the right
#  1->2->3->4->8->6->9 , after rotating for 2 times Cecomes , 8->6->9->1->2->3->4->

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(8)
node5 = Node(6)
node6 = Node(9)

head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node5)
node5.setNext(node6)

print('LinkedList1:')
traverse(head)

#finding length of linkedlist
def len_ll(head):
  count=0
  temp=head
  while temp:
    temp=temp.getNext()
    count+=1

  return count

def rotate_right(head,k):
  k = k % len_ll(head)
  for i in range(k):
    temp=head
    while temp.getNext().getNext():
      temp=temp.getNext()
    tail = temp.getNext()
    temp.setNext(None)
    tail.setNext(head)
    head=tail
  return head


head = rotate_right(head,3)
traverse(head)
```

```python
#Q10. Clone a linked list with next and random pointer

# Given a linked list of size N where each node has two links: one pointer points to the next node and the
# second pointer points to any node in the list. The task is to create a clone of this linked list in O(N) time.


# Note: The pointer pointing to the next node is ‘next‘ pointer and the one pointing to an arCitrary node is
# called ‘arCit’ pointer as it can point to any arCitrary node in the linked list.


class Node_r:
  def __init__(self,data,next=None,arbit=None):
    self.data=data
    self.next=next
    self.arbit=arbit

  def getData(self):
    return self.data

  def setData(self,data):
    self.data=data

  def getNext(self):
    return self.next

  def setNext(self,node):
    self.next=node

  def getArbit(self):
    return self.arbit

  def setArbit(self,node):
    self.arbit=node


def traverse(head):
  temp=head
  while temp:
    print(temp.getData(),end='->')
    temp=temp.getNext()
  print()


# traverse path of arbit pointers
def traverse_arbit(head):
  temp=head
  temp_r=temp.getArbit()
  while temp:
    print(temp.getData(),end='->')
    if temp_r:
      print(temp_r.getData(),end='->')
      print('x',end='->')
    else:
      print('x',end='->')

    temp=temp.getNext()
    if temp:
      temp_r=temp.getArbit()
  print()


head = Node_r(1)
node1 = Node_r(2)
node2= Node_r(3)
node3= Node_r(4)
node4= Node_r(5)


head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)


head.setArbit(node2)
node1.setArbit(head)
node2.setArbit(node4)
node3.setArbit(node2)
node4.setArbit(node1)


print("LinkedList Traverse Path")
traverse(head)
print()
print("Arbitary pointers traverse path")
traverse_arbit(head)
print()


def create_copy_ll(head):

  def insert_copies(head):
    temp=head
    copyhead=None
    while temp:
      if not copyhead:
        copyhead = Node_r(temp.getData())
        node=copyhead
      else:
        node = Node_r(temp.getData())

      next_node = temp.getNext()
      temp.setNext(node)
      node.setNext(next_node)
      temp=next_node

    return head,copyhead


  def assign_arbit_tocopy(head):
    temp=head
    while temp:
      temp.getNext().setArbit(temp.getArbit().getNext())
      temp=temp.getNext().getNext()


  def break_original_and_copy(head):
    temp=head
    while temp:
      original_next_node=temp.getNext().getNext()
      if temp.getNext().getNext():
        temp.getNext().setNext(temp.getNext().getNext().getNext())
      temp.setNext(original_next_node)
      temp=original_next_node

    return head

  head,copyhead = insert_copies(head)
  assign_arbit_tocopy(head)
  head = break_original_and_copy(head)

  return head,copyhead


head,copyhead = create_copy_ll(head)


print('Copied LinkedList')
traverse(copyhead)

print('Copied Arbitary pointers traverse path')
traverse_arbit(copyhead)
```
